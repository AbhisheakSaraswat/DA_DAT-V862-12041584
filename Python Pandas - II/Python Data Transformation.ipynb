### Pandas Pivot Table:

We can summarize the data using pivot table with aggregate functions. 

<b> Note : </b> Pandas pivot tabale by default give you the average of the data.

```python
newdf.head(2)
```

```python
pd.pivot_table(data = newdf, values = 'sales', index = "sub-category")
```

```python
pivot = pd.pivot_table(data = newdf, values = 'sales', index = "sub-category",
                      aggfunc=['sum','mean','count'])
```

```python
pivot
```

```python
pivot = pd.pivot_table(data = newdf, values = ['sales',"profit","quantity"],
                       index = "sub-category",
                      aggfunc= {'sales' : 'sum','profit' : 'sum', 'quantity' : ['sum','mean']})
```

```python
pivot
```

### Pandas loc and iloc:

* loc - > location
* iloc -> index location

<b> loc :</b> I will be follow the original source of current dataframe,
    if it is in number then I will take only numbers or if it is
    string|object then I will taking only string object.
    
_____

<b> iloc :</b> Does not matter for me your original source, I will be follow only the index of dataframe,
if dataframe index is coming in object or string then also I will be follow the index only.
